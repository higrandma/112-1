```python
#5.我所居住的新北市分別有哪些國籍、多少人結婚？
#6.最大值是哪一個呢？

row1 = df.loc[0]
row2 = df.loc[1]

row1_numeric = pd.to_numeric(row1, errors = 'coerce').fillna(0)
row2_numeric = pd.to_numeric(row2, errors = 'coerce' ).fillna(0)

row_sum = row1_numeric + row2_numeric
print(row_sum)

#找最大值
max_col = row_sum.idxmax()
max_row = df[max_col]
max_val = max_row.max()

print("------------------------------------------------")
print("新北市結婚人口中外籍人數第一多")
print(f"國籍:{max_col}，人數:{max_val}")
```

```python
#10.新北市和台北市男女生結婚的人數各相差多少？

相減結果 = []
相減結果2 = []

for i in range(len(新北男)):
    相減結果.append(新北男[i] - 新北女[i])
    
for e in range(len(臺北男)):
    相減結果2.append(臺北男[e] - 臺北女[e])
    
result = sum(相減結果)
result2 = sum(相減結果2)

print('新北市男生結婚人數比女生多',result,'人')
print('臺北市男生結婚人數比女生多',result2,'人')

#兩個縣市的男生結婚人數都比女生少
```
