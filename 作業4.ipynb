```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
```

```python
# 執行1~4頁
newslist = []

for page in range(1,5):
    #連結網站
    response = requests.get("https://www.inside.com.tw/search/教育?page=" + str(page))
    # HTML原始碼解析
    soup = BeautifulSoup(response.text,"html.parser")
    # 取得所有class為post_title的 <h3>
    titles = soup.find_all("h3",{"class":"post_title"})
    
    print(f"================第{str(page)}頁===================")
    
    for title in titles:
        print(title.select_one("a").getText()) #取得標題文字
        
        newstitle=title.select_one("a").getText()
        newslist.append(newstitle)
```

```python
print(newslist)
```

```python
import json

# 存成 json檔並下載

with open('educationlist.json', 'w', encoding='utf-8') as file:
    json.dump(newslist, file, ensure_ascii=False)  
```

```python
# 開啟檔案

with open('educationlist.json', 'r', encoding='utf-8') as file:
    
    # 將JSON文件讀取為Python dict物件
    
    data = json.load(file)
data
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter 
```

```python
data_str = ''.join(data)
data_str
```

```python
#generate a word cloud image
jieba.load_userdict
tags = jieba.analyse.extract_tags(data_str, topK=25, withWeight=True)

for tag in tags:
    print('word:',tag[0],'tf-idf:',tag[1])
```

```python
font_path = '標楷體.ttc'
pngfile = "cloud.jpg"
cloud_mask = np.array(Image.open(pngfile))
```

```python
#draw
wordcloud = WordCloud(font_path=font_path,background_color="white", mask = cloud_mask,contour_width=5,contour_color = 'steelblue').generate(data_str)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show(wordcloud)
```
