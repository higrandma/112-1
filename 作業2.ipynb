```python
data_dict = {}
for i in range(len(df)):
    row = df.iloc[i]
    city_gender = str(row[0]) + str(row[1])  # 取得城市與性別
    data = list(row[2:])          # 取得資料
    data = [int(x.replace(',', '')) if x != '－' else 0 for x in data]  # 將資料轉成 int
    data_dict[city_gender] = data

# 輸出結果
for key, value in data_dict.items():
    print(key, '=', value)
```

```python
import matplotlib.pyplot as plt

# 設定縣市、男女結婚人數
city_data = [    ("New Taipei City", 2080, 2136),    ("Taipei City", 1265, 1269),    ("Taoyuan City", 1309, 1327),    ("Taichung City", 1533, 1549),    ("Tainan City", 883, 891),    ("Kaohsiung City", 1389, 1419),    ("Yilan City", 211, 225),    ("Hsinchu Country", 320, 332),     ("Miaoli City", 247, 245),    ("Changhua City", 520, 530),    ("Nantou City", 215, 209),    ("Yunlin Country ", 283, 291),    ("Jiayi Country", 194, 198),    ("Pingtung Country", 357, 371),    ("Taitung Country", 116, 122),    ("Hualian Country", 185, 65),    ("Penghu Country", 61, 245),    ("Keelung City", 160, 164),    ("Hsinchu City", 252, 260),    ("Kinmen Country", 55, 55),    ("Lianjiang Country", 6, 8),]

# 設定字體大小
plt.rcParams.update({'font.size': 8}) 

# 將各城市的人口數與面積分別存入兩個串列
city_names = []
male_populations = []
female_populations = []
total_populations = []

for data in city_data:
    city_names.append(data[0])
    male_populations.append(data[1])
    female_populations.append(data[2])
    total_populations.append(data[1]+data[2])

# 指定圖表大小
plt.figure(figsize=(30, 15))

# 繪製男女結婚人口數的兩個長條圖
bar_width = 0.4
plt.bar(range(len(city_names)), male_populations, width=bar_width, color='b', label='Male')
plt.bar([x + bar_width for x in range(len(city_names))], female_populations, width=bar_width, color='r', label='Female')

# 繪製總人口數的折線圖
plt.plot(range(len(city_names)), total_populations, color='g', linewidth=2.5, linestyle='--', label='Total')

# 設定x軸的刻度標籤為城市名稱
plt.xticks([x + bar_width/2 for x in range(len(city_names))], city_names)

# 添加圖例、圖表標題與x軸標籤、y軸標籤
plt.legend()
plt.title("Population of Taiwan Cities")
plt.xlabel("City")
plt.ylabel("Population")

# 顯示圖表
plt.show()
```
