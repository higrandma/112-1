```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter # 次數統計
```

```python
dictfile = "字典_utf8.txt"
stopfile = "stopwords.txt"
fontpath = "標楷體.ttc"

mdfile0= "林書豪.txt"
mdfile1 = "戴資穎.txt"
mdfile2 = "王冠閎.txt"
mdfile3 = "郭婞淳.txt"

pngfile = "Taiwan"
```

```python
turkey_mask = np.array(Image.open(pngfile))

jieba.set_dictionary(dictfile)
jieba.analyse.set_stop_words(stopfile)

text0 = open(mdfile0,"r",encoding = "utf-8").read()
text1 = open(mdfile1,"r",encoding = "utf-8").read()
text2 = open(mdfile2,"r",encoding = "utf-8").read()
text3 = open(mdfile3,"r",encoding = "utf-8").read()
```

```python
tags = jieba.analyse.extract_tags(text0, topK = 25)

seg_list = jieba.lcut(text0, cut_all = False)
dictionary = Counter(seg_list)
```

```python
tags1 = jieba.analyse.extract_tags(text1, topK = 25)

seg_list1 = jieba.lcut(text1, cut_all = False)
dictionary1 = Counter(seg_list1)
```

```python
tags2 = jieba.analyse.extract_tags(text2, topK = 25)

seg_list2 = jieba.lcut(text2, cut_all = False)
dictionary2 = Counter(seg_list2)
```

```python
tags3 = jieba.analyse.extract_tags(text3, topK = 25)

seg_list3 = jieba.lcut(text3, cut_all = False)
dictionary3 = Counter(seg_list3)
```

```python
freq = {}
for ele in dictionary:
    if ele in tags:
        freq[ele] = dictionary[ele]
print(freq)
```

```python
freq1 = {}
for a in dictionary1:
    if a in tags1:
        freq1[a] = dictionary1[a]
print(freq1)
```

```python
freq2 = {}
for b in dictionary2:
    if b in tags2:
        freq2[b] = dictionary2[b]
print(freq2)
```

```python
freq3 = {}
for b in dictionary3:
    if b in tags3:
        freq3[b] = dictionary3[b]
print(freq3)
```

```python
wordcloud = WordCloud(background_color = "white", mask=turkey_mask, contour_width = 3, contour_color = 'red', font_path = fontpath).generate_from_frequencies(freq)
plt.figure()
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
