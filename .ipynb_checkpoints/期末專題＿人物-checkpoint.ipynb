```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter # 次數統計
```

```python
import codecs

input_file = "0514土耳其報導.txt"
output_file = "0514土耳其報導_utf8.txt"

with codecs.open(input_file, "r", encoding="Big5") as file:
    content = file.read()

with codecs.open(output_file, "w", encoding="utf-8") as file:
    file.write(content)
```

```python
input_file = "字典.txt"
output_file = "字典_utf8.txt"

with codecs.open(input_file, "r", encoding="Big5") as file:
    content = file.read()

with codecs.open(output_file, "w", encoding="utf-8") as file:
    file.write(content)
```

```python
dictfile = "字典_utf8.txt"
stopfile = "stopwords.txt"
fontpath = "標楷體.ttc"

mdfile = "0514土耳其報導_utf8.txt"
mdfile1 = "0530土耳其報導_utf8.txt"
pngfile = "土耳其國旗.jpg"
```

```python
turkey_mask = np.array(Image.open(pngfile))

jieba.set_dictionary(dictfile)
jieba.analyse.set_stop_words(stopfile)

text = open(mdfile,"r",encoding = "utf-8").read()

text1 = open(mdfile1,"r",encoding = "utf-8").read()
```

```python
tags = jieba.analyse.extract_tags(text, topK = 25)

seg_list = jieba.lcut(text, cut_all = False)
dictionary = Counter(seg_list)
```

```python
dictionary
```

```python
tags1 = jieba.analyse.extract_tags(text1, topK = 25)

seg_list1 = jieba.lcut(text1, cut_all = False)
dictionary1 = Counter(seg_list1)
```

```python
dictionary1
```

```python
freq = {}
for ele in dictionary:
    if ele in tags:
        freq[ele] = dictionary[ele]
print(freq)
```

```python
freq1 = {}
for a in dictionary1:
    if a in tags1:
        freq1[a] = dictionary1[a]
print(freq1)
```

```python
wordcloud = WordCloud(background_color = "white", mask=turkey_mask, contour_width = 3, contour_color = 'red', font_path = fontpath).generate_from_frequencies(freq)
plt.figure()
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(background_color = "white", mask=turkey_mask, contour_width = 3, contour_color = 'red', font_path = fontpath).generate_from_frequencies(freq1)
plt.figure()
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
